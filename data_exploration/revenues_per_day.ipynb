Check if id is primary key:
yes

```python
dupe_pks = (revenues_per_day
    .groupby('id')
    .agg(count=('id', 'size'))
    .query('count > 1')
    .index
    .tolist()
)
dupe_pks
```

Check if title is unique

```python
dupe_pks = (revenues_per_day
    .groupby('title')
    .agg(count=('title', 'size'))
    .query('count >  1')
    .index
    .tolist()
)
dupe_pks
```

Check if day and title can be a business key.

```python
dupe_pks = (revenues_per_day
    .groupby(['title','date'])
    .size()
    .where(lambda x: x > 1)
    .dropna()
    .index
    .tolist()
)
dupe_pks
```
